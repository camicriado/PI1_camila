Vemos que el unico dato que nos podria interesar es el nombre de la coleccion. 
Por lo tanto, procedemos a obtener el valor de name y lo agregamos a una nueva columna llamada collection_name. En caso que no haya datos se aclarara "No hay datos"

```python
def get_collection_name(collection):
    if pd.isna(collection):
        return "No hay datos"
    else:
        try:
            collection_dict = eval(collection)
            if "name" in collection_dict:
                return collection_dict["name"]
            else:
                return "No hay datos"
        except (SyntaxError, TypeError):
            return "No hay datos"


df["collection_name"] = df["belongs_to_collection"].apply(get_collection_name)

```

```python
df.collection_name   #Verificamos que se haya hecho lo correcto. 
```

```python
# Ahora Eliminamos Belong_to_collection.

df.drop("belongs_to_collection", axis=1, inplace=True)
```

**Lo que nos va a interesar es tener una lista con el nombre de todos los Generos de la Pelicula, por lo tanto vamos a extraer estos datos de la columna y los almacenamos en una nueva columna que se llama Genre_Name**

```python


def extract_genre_name(genres):
    try:
        genres_list = ast.literal_eval(genres)
        genre_names = [genre_dict["name"] for genre_dict in genres_list]
        return genre_names
    except (SyntaxError, ValueError):
        return []

df["genre_name"] = df["genres"].apply(extract_genre_name)
```

**Como puede resultarnos util tambien tener el genre_id, tmb extramos este dato y lo almacenamos en una nueva columna**

```python


def extract_genre_id(genres):
    try:
        genres_list = ast.literal_eval(genres)
        genre_id = [genre_dict["id"] for genre_dict in genres_list]
        return genre_id
    except (SyntaxError, ValueError):
        return []

df["genre_id"] = df["genres"].apply(extract_genre_id)
```

**PUNTO 4**

*De haber fechas, deberán tener el formato AAAA-mm-dd, además deberán crear la columna release_year donde extraerán el año de la fecha de estreno.*

Como Bonus, en este punto además vamos a agregar 2 columnas mas: release_month y release_day, ya que luego serán utilizadas en las consultas

```python
df.release_date
```

```python

# Convertir la columna 'release_date' a tipo datetime
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')

# Transformar el formato de la columna 'release_date' a 'AAA-mm-dd'
df['release_date'] = df['release_date'].dt.strftime('%Y-%m-%d')
```

```python
#verifico
df.release_date
```

```python


df['release_date'] = pd.to_datetime(df['release_date'])


# Diccionarios para mapear los nombres de los meses y los días de la semana en español
meses = {
    1: 'Enero',
    2: 'Febrero',
    3: 'Marzo',
    4: 'Abril',
    5: 'Mayo',
    6: 'Junio',
    7: 'Julio',
    8: 'Agosto',
    9: 'Septiembre',
    10: 'Octubre',
    11: 'Noviembre',
    12: 'Diciembre'
}

dias_semana = {
    0: 'Lunes',
    1: 'Martes',
    2: 'Miércoles',
    3: 'Jueves',
    4: 'Viernes',
    5: 'Sábado',
    6: 'Domingo'
}

# Obtener el año, mes y día de la fecha de lanzamiento
df['release_year'] = df['release_date'].dt.year.astype(int)
df['release_month'] = df['release_date'].dt.month.map(meses)
df['release_day'] = df['release_date'].dt.weekday.map(dias_semana)
```

**PUNTO 4**

Crear la columna con el retorno de inversión, llamada return con los campos revenue y budget, dividiendo estas dos últimas revenue / budget, cuando no hay datos disponibles para calcularlo, deberá tomar el valor 0

```python
df['return'] = df.apply(lambda row: row['revenue'] / row['budget'] if row['budget'] != 0 and row['revenue'] != 0 else 0, axis=1)
```

Por último, vamos a importar el dataset de credits y extraer el nombre de los actores y el director de las peliculas. Luego vamos a hacer un Merge con los dos datasets y quedara un DATASET listo para ser utilizado con el EDA

```python

ruta_archivo = os.path.join('Dataset', 'credits.csv')    #Lo descargamos con ruta relativa a la carpeta actual
df_credits = pd.read_csv(ruta_archivo)
```

```python


# Convertir la columna "cast" de cadena de texto a lista de diccionarios
df_credits['cast'] = df_credits['cast'].apply(lambda x: ast.literal_eval(x))

# Definir una función lambda para extraer los valores de la clave "name"
get_names = lambda x: [d['name'] for d in x]

# Aplicar la función lambda a la columna "cast" y asignar los resultados a la columna "Actors"
df_credits['Actors'] = df_credits['cast'].apply(get_names)
```

```python


# Convertir la columna "crew" de cadena de texto a lista de diccionarios
df_credits['crew'] = df_credits['crew'].apply(ast.literal_eval)

# Extraer el director
df_credits['Director'] = df_credits['crew'].apply(lambda row: row[0]['name'] if row else 'No director found')

# Verificar el resultado
print(df_credits['Director'])
```

```python
df_credits['id'] = df_credits['id'].astype(str)  # Convertir la columna 'id' a tipo cadena

df = pd.concat([df, df_credits[['id', 'Actors', 'Director']]], axis=1)

```
